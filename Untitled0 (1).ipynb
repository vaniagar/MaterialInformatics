```python
# Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Dataset
mechanical_data = pd.read_csv('mechanical_properties.csv')
metadata = pd.read_csv('metadata.csv')

# Display the First Few Rows of the Dataset
print("Mechanical Properties Data:")
print(mechanical_data.head())

print("\nMetadata:")
print(metadata.head())

# Data Cleaning
# Convert Strain Rate to Float (if units are inconsistent)
mechanical_data['Strain_Rate_per_second'] = mechanical_data['Strain_Rate_per_second'].apply(
    lambda x: x if isinstance(x, float) else float(str(x).replace('s⁻¹', ''))
)

# Handle Missing Values
# Replace Missing Values in YS and UTS with Mean
mechanical_data['Yield_Strength_MPa'] = mechanical_data['Yield_Strength_MPa'].fillna(mechanical_data['Yield_Strength_MPa'].mean())
mechanical_data['Ultimate_Tensile_Strength_MPa'] = mechanical_data['Ultimate_Tensile_Strength_MPa'].fillna(mechanical_data['Ultimate_Tensile_Strength_MPa'].mean())

# Forward-Fill Missing Test Temperatures
mechanical_data['Test_Temperature_C'] = mechanical_data['Test_Temperature_C'].fillna(method='ffill')

# Remove Rows with Missing Metadata (if critical fields are missing)
cleaned_data = mechanical_data.dropna()

# Statistical Summary of the Cleaned Data
print("\nStatistical Summary:")
print(cleaned_data.describe())

# Visualization 1: Histogram of Yield Strength
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['Yield_Strength_MPa'], kde=True, bins=10, color='skyblue')
plt.title('Histogram of Yield Strength')
plt.xlabel('Yield Strength (MPa)')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Scatter Plot - Test Temperature vs UTS
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Test_Temperature_C', y='Ultimate_Tensile_Strength_MPa', data=cleaned_data, color='orange')
plt.title('Test Temperature vs Ultimate Tensile Strength')
plt.xlabel('Test Temperature (°C)')
plt.ylabel('UTS (MPa)')
plt.show()

# Visualization 3: Pair Plot of Mechanical Properties
sns.pairplot(cleaned_data, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Mechanical Properties', y=1.02)
plt.show()

# Save Cleaned Data to a CSV File (Optional)
cleaned_data.to_csv('cleaned_mechanical_properties.csv', index=False)

```
